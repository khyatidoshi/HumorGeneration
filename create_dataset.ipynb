```python
import numpy as np
import pandas as pd
```

```python
aws_dataset = pd.read_csv('aws_data.csv')
aws_dataset.head()
```

```python
short_jokes = pd.read_csv('shortjokes.csv')
short_jokes.drop(['ID'],axis=1,inplace=True)
short_jokes.head()
```

```python
joke = pd.DataFrame(aws_dataset['question'] + aws_dataset['answer'],columns = ['Joke'])
print(joke)
```

```python
new_dataset = pd.concat([short_jokes,joke],ignore_index=True)
new_dataset
```

```python
new_dataset.to_csv('new_joke_data.csv',index="True")
```
